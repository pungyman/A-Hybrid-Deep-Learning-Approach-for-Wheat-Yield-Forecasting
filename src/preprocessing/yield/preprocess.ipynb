```python
import os
import numpy as np
import pandas as pd

yield_data = pd.read_csv('yield_data.csv')

required_states = ['Uttar Pradesh', 'Madhya Pradesh', 'Rajasthan', 'Haryana', 'Punjab']

yield_data = yield_data[
    (yield_data['Season'] == 'Rabi') & 
    (yield_data['State'].isin(required_states))
]

yield_data = yield_data.drop('Season', axis=1)
```

```python
# Remove rows for years 1998-99 and 1999-00
yield_data = yield_data[~yield_data['Year'].isin(['1998-99', '1999-00'])]
```

```python
yield_data.to_csv('yield_data_5_states.csv', index=False)
```
