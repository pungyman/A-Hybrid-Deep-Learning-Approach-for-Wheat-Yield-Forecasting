```python
import numpy as np
import pandas as pd

yield_data = pd.read_csv('yield_data_5_states.csv')
```

```python
yield_data['derived_yield'] = (yield_data['Production'] / yield_data['Area']) * 1000
```

```python
# Drop the 'Crop' column
yield_data.drop(columns=['Crop'], inplace=True)
```

```python
# Count the number of times derived_yield is greater than Yield and less than Yield
derived_yield_greater_than_yield = len(yield_data[yield_data['derived_yield'] > yield_data['Yield']])
derived_yield_less_than_yield = len(yield_data[yield_data['derived_yield'] < yield_data['Yield']])
```

```python
derived_yield_greater_than_yield, derived_yield_less_than_yield
```

```python
# If the mismatch was due to the Area column being Total Area instead of Area under Cultivation,
# then derived_yield would've been consistently lower than Yield. This is not the case.
```

```python
# Look at the rows with the largest differences between derived_yield and Yield
yield_data['difference'] = yield_data['derived_yield'] - yield_data['Yield']
```

```python
# The yield data likely does not match the derived yield because of rounding errors.
# Both the Area and Production columns are rounded to 2 decimal places.
```

```python
yield_data['district_id'] = yield_data['State'] + '_' + yield_data['District']
```

```python
len(yield_data['district_id'].unique())
```

```python
# we have yield data for 205 districts in our 5 states of interest. (for at least 1 year)
```

```python
# how many years of data do we have for each district
# also for each district what is the start year and end year
yield_data_availability = yield_data.dropna().groupby('district_id')['Year'].agg(['count', 'min', 'max']).reset_index()
yield_data_availability = yield_data_availability.rename(columns={'count': 'num_years', 'min': 'first_year', 'max': 'last_year'})
```

```python
# for every district, the last year is 2023-24. so we can conclude that no district was 'delisted'. 
# i.e., no district disappeared out of existence
```

```python
# also, Rajasthan Pratapgarh is the only district with a broken sequence.
# will investigate
```

```python
missing_data_rows = yield_data[yield_data.isnull().any(axis=1)]
```
